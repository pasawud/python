# edge


```python

import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('baifern_face.jpg')

plt.imshow(img)
```

```python
edges = cv2.Canny(img,127,127)
plt.imshow(edges)
```

```python
edges = cv2.Canny(img,255,0)
plt.imshow(edges)
```

```python
med = np.median(img)
lower = int(max(0,0.7*med))
upper = int(min(255,1.3*med))

edges = cv2.Canny(img,lower,upper)
plt.imshow(edges)
```

```python
blurred_img = cv2.blur(img,(5,5))

edges = cv2.Canny(blurred_img,lower,upper)

plt.imshow(edges)
```

#  contour

```python

import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('internal_external.png',0)
plt.imshow(img,cmap='gray')
```

```python
contours, hierarchy = cv2.findContours(img,cv2.RETR_CCOMP,cv2.CHAIN_APPROX_SIMPLE)
```

#  take external contour


```python
external_contours =  np.zeros(img.shape)

for i in range(len(contours)):
    if hierarchy [0][i][3] == -1:
        cv2.drawContours(external_contours,contours,i,255,-1)
plt.imshow(external_contours, cmap='gray')
    
```
